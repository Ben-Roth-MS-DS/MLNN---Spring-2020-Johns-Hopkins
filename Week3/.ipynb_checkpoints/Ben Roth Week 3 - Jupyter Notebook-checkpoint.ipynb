## 1. Create a 4 dimensional data set with 64 elements and show 2D plots of the data $x_1 \rightarrow y, x_2 \rightarrow y$, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 64
x = np.linspace(0, 1, n) + np.random.rand(4, n)
x = np.vstack([x, np.ones(len(x.T))]).T
y = np.linspace(0, 1, n) + np.random.rand(n) - 1 
```

```python
cols = ['green', 'm', 'red', 'c']

fig, axes = plt.subplots(len(x.T) - 1, 1, figsize = (12, 12))
for i in range(len(x.T) - 1):
    axes[i].set_title('Y vs. x' + str(i))
    axes[i].scatter(x = x.T[i], y = y, c = cols[i])
    
fig.tight_layout()
```

## 2.  Create a model to fit the data. Hint: follow the example from Lesson 3

Following the below equation:

$\beta = (X^T X)^{-1} Y^T X$

```python
left = np.linalg.inv(np.dot(x.T, x))
right = right = np.dot(y.T, x)
beta = np.dot(left, right)
pred = np.dot(x, beta)
```

## 3. Plot the model's prediction in 2D for 2 of the dimensions ($x_1 \rightarrow y_p, x_2 \rightarrow y_p$) along with the original points

```python
plt.title('Predictions and Ys for X1')
plt.scatter(x.T[0], pred, c = 'purple')
plt.scatter(x.T[0], y, c = 'green')
```

```python
plt.title('Predictions and Ys for X2')
plt.scatter(x.T[0], pred, c = 'c')
plt.scatter(x.T[0], y, c = 'm')
```

## 4. Read in `mlnn/data/Credit.csv` with Pandas and create a model to predict Credit Rating (`Rating`). Use only the numeric columns in your model, but feel free to experiment which which columns you believe are better predicters of Credit Rating

```python
import pandas as pd
credit = pd.read_csv('../data/Credit.csv')
credit.head()
```

```python
X = credit[['Income', 'Limit', 'Age', 'Balance']].as_matrix()

X = np.vstack([X.T, np.ones(len(X))]).T
X
```

```python
y = credit['Rating']
```

```python
left = np.linalg.inv(np.dot(X.T, X))
right = right = np.dot(y.T, X)
beta = np.dot(left, right)
pred = np.dot(X, beta)
```

### 5. Plot your results (Bonus if you use 3D plots). Show as many of your columns vs. credit rating that you can. 

```python
#2d plots
fig, axes = plt.subplots(len(X.T) - 1, 1, figsize = (12, 12))
for i in range(len(X.T) - 1):
    axes[i].set_title('Predicted Credit Rating and Actual Credit Rating Against ' + credit[['Income', 'Limit', 'Age', 'Balance']].columns[i] 
                      + ' (green = true, purple = prediction)')
    axes[i].scatter(x = X.T[i], y = y, c = 'green')
    axes[i].scatter(x = X.T[i], y = pred, c = 'purple')
    
fig.tight_layout()
```

```python
#3d plots
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.title('X = Income, Y = Limit, Z = Predicted Credit Rating (red) & Actual Credit Rating (blue)')
ax.view_init(23, 30)
ax.scatter(X.T[0], X.T[1], pred.T, zdir='z', c='r')
ax.scatter(X.T[0], X.T[1], y.T, zdir='z', c='b')
```
