### And here we read in the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
bikes = pd.read_csv('../data/2016-Q1-cabi-trip-history-data.csv')
bikes.head()
bikes['start'] = pd.to_datetime(bikes['Start date'], infer_datetime_format=True)
bikes['end'] = pd.to_datetime(bikes['End date'], infer_datetime_format=True)

bikes.head()
```

### Create a new column that represents the hour of the day 

```python
bikes['hour_of_day'] = (bikes.start.dt.hour + (bikes.start.dt.minute/60).round(2))
```

### Aggregate to get a count per hour/minute of the day across all trips

```python
hours = bikes.groupby('hour_of_day').agg('count')
hours['hour'] = hours.index

hours.start.plot()
# import seaborn as sns

# sns.lmplot(x='hour', y='start', data=hours, aspect=1.5, scatter_kws={'alpha':0.2})
```

## 1. Create several models fit to `hour_of_day` with various polynomial degrees across different time periods. i.e. fit a model for the early bump, fit another model for the second bump, etc. 

```python
#X = partial_hours[['hour']]

#define bumps
first_bump = hours.loc[5.5:9]

X = first_bump[['hour']]

```

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures


#define bumps and X,y
first_bump = hours.loc[5:10]
X1st = first_bump[['hour']]
y1st = first_bump.start

second_bump = hours.loc[10:15]
X2nd = second_bump[['hour']]
y2nd = second_bump.start

third_bump = hours.loc[15.5:23.5]
X3rd = third_bump[['hour']]
y3rd = third_bump.start


#poly = 2 for second bump
poly4 = PolynomialFeatures(degree=4)

#poly = 10 for first bump
poly10 = PolynomialFeatures(degree=10)

#poly = 6 for third bump
poly6 = PolynomialFeatures(degree=6)


#polynomially transform each set of bumps
x1st_10 = poly10.fit_transform(X1st)
x2nd_4 = poly4.fit_transform(X2nd)
x3rd_6 = poly6.fit_transform(X3rd)

#fit models
linear1st = linear_model.LinearRegression()
linear1st.fit(x1st_10, y1st)

linear2nd = linear_model.LinearRegression()
linear2nd.fit(x2nd_4, y2nd)

linear3rd = linear_model.LinearRegression()
linear3rd.fit(x3rd_6, y3rd)
```

```python
#plot results
import numpy as np
from matplotlib.pyplot import figure
#increase plot size
figure(num=None, figsize=(8, 6))
plt.scatter(hours.hour, hours.start, c = 'b')
plt.xlabel('hours')
plt.ylabel('count')
plt.plot(first_bump.hour, np.dot(x1st_10, linear1st.coef_) + linear1st.intercept_, c='g', linewidth = 5)
plt.plot(second_bump.hour, np.dot(x2nd_4, linear2nd.coef_) + linear2nd.intercept_, c='r', linewidth = 5)
plt.plot(third_bump.hour, np.dot(x3rd_6, linear3rd.coef_) + linear3rd.intercept_, c='c', linewidth = 5)
```

## 2. Using the previous models, create new models fit to `hour_of_day` with 3 Ridge Regression $\alpha$ (alpha) Ridge Coefficient values
. Plot all the models predictions in one chart

```python
#create and fit ridge models
ridge1st = linear_model.Ridge(alpha = 4)
ridge2nd = linear_model.Ridge(alpha = 4)
ridge3rd = linear_model.Ridge(alpha = 4)



ridge1st.fit(x1st_10, y1st)
ridge2nd.fit(x2nd_4, y2nd)
ridge3rd.fit(x3rd_6, y3rd)

#increase plot size
figure(num=None, figsize=(8, 6))

#plot original data, polynomial non-ridged data, and polynomial Ridge data to compare results
plt.scatter(hours.hour, hours.start, c = 'b')
plt.xlabel('hours')
plt.ylabel('count')
plt.legend(loc="upper left")

plt.plot(first_bump.hour, np.dot(x1st_10, ridge1st.coef_) + ridge1st.intercept_, c='g', linewidth = 5)
plt.plot(second_bump.hour, np.dot(x2nd_4, ridge2nd.coef_) + ridge2nd.intercept_, c='g', linewidth = 5)
plt.plot(third_bump.hour, np.dot(x3rd_6, ridge3rd.coef_) + ridge3rd.intercept_, c='g', linewidth = 5)

```
